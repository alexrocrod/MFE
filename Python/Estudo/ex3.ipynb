```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## ex3
## metodo tranformacao mas com 2 ramos:

n_pnts = int(1e5)

u = np.random.rand(n_pnts)


## Aplicar a trasformação
x = np.zeros(n_pnts)

for i in range(n_pnts):
    if u[i]<=3/4:
        x[i] = (4/3*u[i])**(1/3)
    else:
        x[i] = ((1-u[i])*4)**(-1/9)
```

```python
## histograma

n_bins = 50
xmax = 2 # depois é praticamente constante

values, bins = np.histogram(x, bins=np.linspace(0, xmax, n_bins))

dx = (xmax - 0) / n_bins
h = values / np.sum(values) / dx
```

```python

n_pnts_init = int(n_pnts*(1-0)/xmax) # pontos antes de x=1 
x_values1 = np.linspace(0, 1, n_pnts_init)
x_values2 = np.linspace(1, xmax, n_pnts-n_pnts_init)

f_teo=[]
f_teo.extend(9/4*x_values1**2)
f_teo.extend(9/4*x_values2**(-10))
x_values = np.linspace(0, xmax, n_pnts)


plt.plot(bins[:-1], h, '*r')
plt.plot(x_values, f_teo, '-b')
plt.show()

```
