```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## ex4

p=0.65

n_pnts = int(1e2)
u = np.random.rand(n_pnts)

## Aplicar a trasformação
x = np.zeros(n_pnts)

for i in range(n_pnts):
    if u[i]<p:
        x[i] = 1
    else:
        x[i] = 0

w=[1-p,p]
```

```python
D = np.zeros(n_pnts)
D[0]= 100
wi = np.ones(n_pnts)
W=0

for i in range(1,n_pnts):
    D[i] = D[i - 1] * 2 * w[int(x[i])]

for i in range(n_pnts):
    wi[i]=wi[i-1]*w[int(x[i])]

W = np.mean(wi)

plt.plot(np.arange(0, n_pnts, 1), D)
plt.show()
print("Dinheiro final: ",D[-1])
print("Dinheiro Maximo: ",np.max(D),", t=",np.argmax(D))
```

```python
Wteo=1 + p*np.log2(p) + (1-p)*np.log2(1-p)
print("W: ",W, ", Wteo: ",Wteo, "  ", W/Wteo)
```
