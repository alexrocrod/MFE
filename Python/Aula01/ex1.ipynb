```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_pnts = int(1e6)

u = np.random.rand(n_pnts)
```

```python
x = np.arccos(1 - 2 * u)
n_bins = 20
values, bins = np.histogram(x, bins=np.linspace(0, np.pi, n_bins))

dx = (np.pi - 0) / n_bins
values = (values / np.sum(values)) / dx

x_values = np.linspace(0, np.pi, 1000)

plt.figure(2)
plt.plot(bins[0:-1], values, '*r')
plt.plot(x_values, np.sin(x_values) / 2, '-b')
plt.xlabel("theta")
plt.ylabel("p(theta)")
```

```python
phi = 2*np.pi*np.random.rand(n_pnts)
x=np.sin(x)*np.cos(phi)
y=np.sin(x)*np.sin(phi)
z=np.cos(x)
```

```python
fig=plt.figure()
ax=plt.axes(projection="3d")
ax.plot3d(x,y,z)
```
