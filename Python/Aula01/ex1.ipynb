```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Alexandre Rodrigues 92993
# MFE - Aula01 - Ex.1

n_pnts = int(1e3)

u = np.random.rand(n_pnts)

n_bins = 20
values, bins = np.histogram(u, bins=np.linspace(0, 1, n_bins))

dx = (1 - 0) / n_bins
values = (values / np.sum(values)) / dx

plt.figure(1)
plt.plot(bins[0:-1], values, '*r')
plt.plot([0, 1], [1, 1], '-b')
plt.xlabel("theta")
plt.ylabel("F-1(u)")

theta = np.arccos(1 - 2 * u)
n_bins = 20
values, bins = np.histogram(theta, bins=np.linspace(0, np.pi, n_bins))

dx = (np.pi - 0) / n_bins
values = (values / np.sum(values)) / dx

dx = 0.001
x_values = np.arange(0, np.pi, dx)

plt.figure(2)
plt.plot(bins[0:-1], values, '*r')
plt.plot(x_values, np.sin(x_values) / 2, '-b')
plt.xlabel("theta")
plt.ylabel("p(theta)")
```

```python
# Exercício 1
# b)

phi = 2 * np.pi * np.random.rand(n_pnts)

x = np.cos(phi) * np.sin(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(theta)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot3D(x, y, z, '.r')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```
